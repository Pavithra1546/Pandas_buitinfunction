# Pandas Buitin function
1)loc[]
  DataFrame.loc[] method is used to retrieve rows from Pandas DataFrame

```python
import pandas as pd
data = pd.read_csv("Hit or Flop movie Dataset.csv",index_col ="Movie Name")
data.loc["Sardar"]

```

# 2)iloc[]
Rows can also be selected by passing integer location to an iloc[] function.

```python
data.iloc[3:5]
```

# 3)head()
head(n) is used to return the first n rows of a dataset

```python
data.head(5)
```

# 4. memory_usage()
memory_usage() returns a Pandas Series having the memory usage of each column (in bytes) in a Pandas DataFrame. By specifying the deep attribute as True, we can get to know the actual space being taken by each column. 

```python
data.memory_usage(deep=True)
```

# 5. to_datetime()
to_datetime() converts a Python object to datetime format. It can take an integer, floating point number, list, Pandas Series, or Pandas DataFrame as argument. to_datetime() is very powerful when the dataset has time series values or dates.

```python
pd.to_datetime(data['Movie Release Date'])
```

# 6) value_counts()
value_counts() returns a Pandas Series containing the counts of unique values. 

```python
data['Genre'].value_counts()
```

# 7) groupby()
groupby() is used to group a Pandas DataFrame by 1 or more columns, and perform some mathematical operation on it.

```python
data.groupby(by='Movie Name').YouTube_Trailer_Views.mean()
```

# 8) sort_values()
sort_values() is used to sort column in a Pandas DataFrame (or a Pandas Series) by values in ascending or descending order. 

```python
data.sort_values(by='Movie Name', inplace=True)
data
```
