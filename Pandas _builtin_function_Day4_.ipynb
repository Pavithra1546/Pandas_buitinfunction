```python
data=pd.DataFrame({"Reg No":[4036,4037,4038,4039],'Nanoscience':[95,80,88,93],"Ethics":[88,97,89,76]})
data
```

```python
import pandas as pd
survey = pd.DataFrame({'James': ['I liked it', 'It could use a bit more salt'], 'Emily': ['It is too sweet', 'Yum!']})
survey
```

# we can specify an index column when creating a dataframe via the 'index' argument.

```python
survey = pd.DataFrame({'James': ['I liked it', 'It could use a bit more salt'], 'Emily': ['It is too sweet', 'Yum!']},
                     index = ['Product A', 'Product B'])
survey
```

# Reset index
# Try playing around with 'drop' and 'inplace' and see what they do


```python
survey.reset_index(drop = True,inplace=True)
survey
```
